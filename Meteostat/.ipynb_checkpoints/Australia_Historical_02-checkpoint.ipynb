```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Reference/Australia/Sydney_1948_2018.csv')
```

```python
data.describe()
```

```python
data = data.rename(columns={'T':'AvgYearlyTemp','TM':'AvgYearlyMaxTemp (C)','Tm':'AvgYearlyMinTemp','PP':'RainSnowTotalAmt','V':'YearlyAvgWindSpeed','RA':'RainDays','SN':'SnowDays','TS':'StormDays','FG':'FoggyDays','TN':'TornadoDays','GR':'HailDays'})
```

```python
reduced_data = data[['Year','AvgYearlyTemp','AvgYearlyMaxTemp (C)','AvgYearlyMinTemp']]
```

```python
reduced_data = data[['Year','AvgYearlyMaxTemp (C)']]
```

```python
df10 = pd.DataFrame(reduced_data)
```

```python
df10.count()
```

```python
australia_df_1948_2018 = df10.dropna(how='any')
```

```python
australia_df_1948_2018.reset_index(drop=True)
```

```python
x_axis = australia_df_1948_2018['Year']
```

```python
y_axis = australia_df_1948_2018['AvgYearlyMaxTemp (C)']
```

```python
fig, ax = plt.subplots()

fig.set_figheight(8)
fig.set_figwidth(15)
ax.plot(x_axis,y_axis)
ax.set_xlabel('Years')
ax.set_ylabel('Temperature(C)')
ax.set_title('Australia')
plt.grid()
```

```python
restricted_australiaghe2 = australia_df_1948_2018.loc[australia_df_1948_2018["Year"] >= 1977, :]
restricted_australiaghe = restricted_australiaghe2.loc[restricted_australiaghe2["Year"] < 2008, :]
```
